```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Actual_Generation_per_Production_Type_modified.xlsx', index_col=[0])
df.head()
```

```python
df['Date and Time']=pd.to_datetime(df['Date and Time'])
```

```python
df['Day of the week'] = df['Date and Time'].dt.day_name()
```

```python
df
```

```python
df['Date and Time'] = df['Date and Time'].dt.round('T')
```

```python
df
```

```python
df.set_index('Date and Time', inplace=True)
```

```python
df.loc['2020-01-01']['Solar'].head(60)
```

```python
df.loc['2020-01-01']['Solar'].max()
```

```python
df.loc['2020-01-01']['Solar'].sum()
```

```python
df.loc['2020-01-01']['Solar'].mean()
```

```python
df['Solar'].resample('D').sum()
```

```python
df.resample('D').max()
```

```python
solar_max = df['Solar'].resample('D').max()
solar_max.plot()
```

```python
windon_max = df['Wind Onshore'].resample('D').max()
windoff_max = df['Wind Offshore'].resample('D').max()
wind_max = windon_max + windoff_max
windon_max.plot()
windoff_max.plot()
wind_max.plot()
```

```python
biomass_max = df['Biomass'].resample('D').max()
biomass_max.plot()
```

```python
hydro_ps_max = df['Hydro Pumped Storage'].resample('D').max()
hydro_rorp_max = df['Hydro Run-of-river and poundage'].resample('D').max()
hydro_wr_max = df['Hydro Water Reservoir'].resample('D').max()
hydro_max = hydro_ps_max + hydro_rorp_max + hydro_wr_max


hydro_ps_max.plot()
hydro_rorp_max.plot()
hydro_wr_max.plot()
hydro_max.plot()
```

```python
wind_solar_mix = wind_max + solar_max

plt.figure(figsize=(18, 6))
plt.title('Total electricity generation in 2020 for Solar and Wind', fontdict = {'fontweight': 'bold', 'fontsize': 18})

plt.fill_between(wind_solar_mix.index, wind_solar_mix, label='Wind and Solar total generation')
plt.fill_between(wind_max.index, wind_max, label='Wind total generation')
plt.fill_between(solar_max.index, solar_max, label='Solar total generation')


plt.xlabel("Time period")
plt.ylabel('Electricity generation')
plt.legend()
plt.show()
```
