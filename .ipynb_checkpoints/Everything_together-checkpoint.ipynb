```python
result = pd.concat([df_generation['Total Generation(TWh))'], df_load['Actual Total Load(TWh)'], total_flow['Germany Import(TWh)']], axis=1)
```

```python
result['Difference'] = result['Total Generation(TWh))'] + result['Germany Import(TWh)'] - result['Actual Total Load(TWh)']
```

```python
plt.figure(figsize=(18, 6))


plt.title('Generation + Import - Load', fontdict = {'fontweight': 'bold', 'fontsize': 18})


plt.plot(result.index , result['Difference'],  label='difference')
plt.plot(result.index , result['Difference']*0,  color='red')


plt.ylabel('Electricity, TWh')
plt.xlabel('Timestep')
plt.legend()

plt.show()
```

```python
df_load_half_first = df_load.loc['2020-01-01':'2020-07-01']
df_load_half_second = df_load.loc['2020-07-01':'2021-01-01']
```

```python
plt.figure(figsize=(18, 8))

plt.subplot(2, 1, 1)
plt.title('Second half of the year', fontdict = {'fontweight': 'bold', 'fontsize': 18})
plt.plot(df_load_half_first.index , df_load_half_first['Actual Total Load(TWh)'],  label='Load')
plt.legend(loc='upper center')

plt.subplot(2, 1, 2)
plt.plot(total_flow.loc['2020-01-01':'2020-07-01'].index , total_flow['Germany Export(TWh)'].loc['2020-01-01':'2020-07-01'],  label='Export')
plt.plot(total_flow.loc['2020-01-01':'2020-07-01'].index , total_flow['Germany Import(TWh)'].loc['2020-01-01':'2020-07-01'],  label='Import')
plt.legend(loc='upper center')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(18, 8))

plt.subplot(2, 1, 1)
plt.title('First half of the year', fontdict = {'fontweight': 'bold', 'fontsize': 18})
plt.plot(total_flow.loc['2020-01-01':'2020-07-01'].index , (total_flow['Germany Export(TWh)'].loc['2020-01-01':'2020-07-01']
                                                            -total_flow['Germany Import(TWh)'].loc['2020-01-01':'2020-07-01']),
                                                             label='Export-Import')
plt.plot(total_flow.loc['2020-01-01':'2020-07-01'].index , total_flow.loc['2020-01-01':'2020-07-01']*0,  color='red')

plt.legend(loc='upper center')

plt.subplot(2, 1, 2)
plt.title('Second half of the year', fontdict = {'fontweight': 'bold', 'fontsize': 18})
plt.plot(total_flow.loc['2020-07-01':'2021-01-01'].index , (total_flow['Germany Export(TWh)'].loc['2020-07-01':'2021-01-01']
                                                            -total_flow['Germany Import(TWh)'].loc['2020-07-01':'2021-01-01']),
                                                             label='Export-Import')
plt.plot(total_flow.loc['2020-07-01':'2021-01-01'].index , total_flow.loc['2020-07-01':'2021-01-01']*0,  color='red')
plt.legend(loc='upper center')

plt.tight_layout()
plt.show()
```
