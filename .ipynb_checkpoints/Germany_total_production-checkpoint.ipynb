```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Global-Electricity-Review-2020-data.xlsx',
                         sheet_name='Data')
df.head()
```

```python
df = df.drop(columns=['Link', 'Source'])
df.rename(columns = { 'Value (TWh)' : 'Value(TWh)'}, inplace = True)
```

```python
germany_filt = (df['Country'] == 'Germany') & (df['Year'] >= 2016)
```

```python
df_germany = df.loc[germany_filt]
df_germany = df_germany.drop(index = df_germany[df_germany['Variable'] == 'Other renewables'].index)
df_germany.reset_index(inplace = True)
```

```python
df_germany.drop(columns = ['index'], inplace = True)
```

```python
df_germany
```

```python
pd.unique(df_germany['Variable'])
```

```python
plt.figure(figsize=(12, 6))

plt.style.use('default')

plt.title('Total Production in 2016, TWh', fontdict = {'fontweight': 'bold', 'fontsize': 18})

types = ['Coal', 'Gas', 'Other fossil', 'Wind', 'Solar', 'Hydro', 'Biomass and waste', 'Nuclear']

year_2016 = df_germany.loc[(df_germany['Year'] == 2016) & (df_germany['Variable'] != 'Net imports') 
                           & (df_germany['Variable'] != 'Production') 
                           & (df_germany['Variable'] != 'Demand')]

total = sum(year_2016['Value(TWh)'])

def absolute_value(val):
    a  = np.round(val/100.*year_2016['Value(TWh)'].sum(), 0)
    return a

plt.pie(year_2016['Value(TWh)'], labels = types, autopct = absolute_value, pctdistance = 0.8)




plt.show()

```
