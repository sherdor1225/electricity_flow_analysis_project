```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Cross_Border_totalflow_cleaned.xlsx', index_col=[0])
df.head()
```

```python
df.tail(10)
```

```python
total_df=df.T.sum(axis=1).to_frame().T
```

```python
total_df.head()
```

```python
df.head(24)
```

```python
df_new = pd.Series(pd.date_range(start='1/1/2020 00:00:00', end='31/12/2020 23:00:00', freq='1h'))
df_new
```

```python
df.reset_index(inplace=True)
```

```python
horizontal_stack = pd.concat([df_new, df], axis=1)
horizontal_stack.drop(columns=['Time'], inplace=True)
horizontal_stack.rename(columns={0: 'Timestep'}, inplace=True)
horizontal_stack['Timestep']=pd.to_datetime(horizontal_stack['Timestep'])
horizontal_stack.set_index('Timestep', inplace=True)
horizontal_stack.head(24)*1000
```

```python
horizontal_stack.tail()
```

```python
hor_stack_day_mean=horizontal_stack.resample('D').mean()*1000
```

```python
hor_stack_day_mean.head()
```

```python
hor_stack_day_im = hor_stack_day_mean.iloc[:, 0:21:2]
hor_stack_day_ex = hor_stack_day_mean.iloc[:, 1:22:2]
```

```python
hor_stack_day_ex_std=horizontal_stack.iloc[:, 1:22:2].resample('D').std()*1000
hor_stack_day_im_std=horizontal_stack.iloc[:, 0:21:2].resample('D').std()*1000
```

```python
hor_stack_day_im_tot = hor_stack_day_mean['Germany Import(TWh)']
hor_stack_day_ex_tot = hor_stack_day_mean['Germany Export(TWh)']
```

```python
hor_stack_day_ex.head()
```

```python
hor_stack_day_im.head()
```

```python
hor_stack_day_im_tot.head()
```

```python
fig_1 = plt.figure(figsize=(18,12))

axes_1 = fig_1.add_axes([0.0, 0.0, 0.9, 0.9])

axes_1.plot(hor_stack_day_ex.index, hor_stack_day_ex['Germany (DE) > Austria (AT)(TWh)'], 'b-', label='mean_DE>AT')
axes_1.fill_between(hor_stack_day_ex.index, hor_stack_day_ex['Germany (DE) > Austria (AT)(TWh)'] - 
                 hor_stack_day_ex_std['Germany (DE) > Austria (AT)(TWh)'], 
                 hor_stack_day_ex['Germany (DE) > Austria (AT)(TWh)'] + 
                 hor_stack_day_ex_std['Germany (DE) > Austria (AT)(TWh)'], color='b', alpha=0.2)

axes_1.set_ylabel('Cross-border flow, GWh')
axes_1.set_title('Cross-border export from DE to AT')
axes_1.legend(loc=0)
sns.set_style("darkgrid")
sns.set_context("paper", font_scale=1.5, rc={"lines.linewidth": 2.5})
```

```python
fig_2 = plt.figure(figsize=(18,12))

axes_2 = fig_2.add_axes([0.0, 0.0, 0.9, 0.9])

axes_2.plot(hor_stack_day_im.index, hor_stack_day_im['Austria (AT) > Germany (DE)(TWh)'], 'r-', label='mean_AT>DE',
           marker ='o', markersize=5, markerfacecolor='red',)
axes_2.fill_between(hor_stack_day_im.index, hor_stack_day_im['Austria (AT) > Germany (DE)(TWh)'] - 
                 hor_stack_day_im_std['Austria (AT) > Germany (DE)(TWh)'], 
                 hor_stack_day_im['Austria (AT) > Germany (DE)(TWh)'] + 
                 hor_stack_day_im_std['Austria (AT) > Germany (DE)(TWh)'], color='r', alpha=0.2)

axes_2.set_ylabel('Cross-border flow, GWh')
axes_2.set_title('Cross-border import from AT to DE')
axes_2.legend(loc=0)
sns.set_style("darkgrid")
sns.set_context("paper", font_scale=1.5, rc={"lines.linewidth": 2.5})
```

```python
fig_3 = plt.figure(figsize=(18,8))

axes_3 = fig_3.add_axes([0.0, 0.0, 0.9, 0.9])

axes_3.plot(hor_stack_day_ex.index, hor_stack_day_ex['Germany (DE) > France (FR)(TWh)'], 'b-', label='mean_DE>FR')
axes_3.fill_between(hor_stack_day_ex.index, hor_stack_day_ex['Germany (DE) > France (FR)(TWh)'] - 
                 hor_stack_day_ex_std['Germany (DE) > France (FR)(TWh)'], 
                 hor_stack_day_ex['Germany (DE) > France (FR)(TWh)'] + 
                 hor_stack_day_ex_std['Germany (DE) > France (FR)(TWh)'], color='b', alpha=0.2)

axes_3.set_ylabel('Cross-border flow, GWh')
axes_3.set_title('Cross-border export from DE to FR')
axes_3.legend(loc=0)
sns.set_style("darkgrid")
sns.set_context("paper", font_scale=1.5, rc={"lines.linewidth": 2.5})
```

```python
fig_4 = plt.figure(figsize=(18,8))

axes_4 = fig_4.add_axes([0.0, 0.0, 0.9, 0.9])

axes_4.plot(hor_stack_day_im.index, hor_stack_day_im['France (FR) > Germany (DE)(TWh)'], 'r-', label='mean_FR>DE',
           marker ='o', markersize=4, markerfacecolor='red', markeredgecolor='black', markeredgewidth=0.5)
axes_4.fill_between(hor_stack_day_im.index, hor_stack_day_im['France (FR) > Germany (DE)(TWh)'] - 
                 hor_stack_day_im_std['France (FR) > Germany (DE)(TWh)'], 
                 hor_stack_day_im['France (FR) > Germany (DE)(TWh)'] + 
                 hor_stack_day_im_std['France (FR) > Germany (DE)(TWh)'], color='r', alpha=0.2)

axes_4.set_ylabel('Cross-border flow, GWh')
axes_4.set_title('Cross-border import from FR to DE')
axes_4.legend(loc=0)
sns.set_style("darkgrid")
sns.set_context("paper", font_scale=1.5, rc={"lines.linewidth": 2.5})
```
