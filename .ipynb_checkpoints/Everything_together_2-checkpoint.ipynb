```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df = pd.read_excel('Cross_Border_totalflow_cleaned.xlsx', index_col=[0])
df.head()
```

```python
df_gener = pd.read_excel('Generation_per_hour_total.xlsx')
df_gener.tail()
```

```python
df_gener['Total Generation(TWh)'] = df_gener.sum(axis=1)
df_gener.head()
```

```python
df_gener.drop(index=[8783])
```

```python
gener_monthly = df_gener.resample('M', on='Date and Time').sum()
```

```python
gener_monthly.index = gener_monthly.index.date
```

```python
gener_monthly.tail()
```

```python
df_load = pd.read_excel('#Load_hourly_modified.xlsx')
df_load.tail()
```

```python
df_load.drop(index=[8783], inplace=True)
```

```python
df_load_monthly = df_load.resample('M', on='Timestep').sum()
```

```python
df_load_monthly.index = df_load_monthly.index.date
```

```python
df_load_monthly.tail()
```

```python
data_daily = df.resample('D').sum()
```

```python
data_monthly = df.resample('M').sum()
```

```python
fig, ax = plt.subplots(figsize=(18, 6))


# define the shift
shift = pd.to_timedelta('5D')
barWidth = 5
# modify the base of each columns, can do with a for loop
ax.bar(data_monthly.index, data_monthly["Germany Export(TWh)"], width=barWidth, label='Total export') 
ax.bar(data_monthly.index + shift, data_monthly["Germany Import(TWh)"], width=barWidth, label='Total import')
#ax.bar(testdata.index - shift, testdata["C"])
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %D'))
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(18, 3))


# define the shift
shift = pd.to_timedelta('5D')
width=5
# modify the base of each columns, can do with a for loop
ax.bar(gener_monthly.index, gener_monthly['Total Generation(TWh)'], width, color='green', label='Total generation') 
ax.bar(df_load_monthly.index + shift, df_load_monthly["Actual Total Load(TWh)"], width, color='red', label='Total load')
#ax.bar(testdata.index - shift, testdata["C"])
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

plt.legend()
plt.show()

```

```python
df_flows = pd.read_excel('Cross_Border_totalflow_cleaned.xlsx', index_col=[0])
df_flows.head()
```

```python
df_flows_monthly = df_flows.resample('M').sum()
```

```python
df_flows_monthly.index = df_flows_monthly.index.date
```

```python
df_outflows_monthly = df_flows_monthly.iloc[:, 1:22:2]
```

```python
df_outflows_monthly.head()
```

```python
df_inflows_monthly = df_flows_monthly.iloc[:, 0:21:2]
```

```python
df_inflows_monthly.head()
```

```python
df_flows_monthly.tail()
```

```python
fig, ax = plt.subplots()

ax = df_inflows_monthly.plot.bar(stacked=True, figsize=(16,8), ax=ax, cmap='tab20')

ax.set_xticklabels([x.strftime("%Y-%m") for x in df_inflows_monthly.index])

ax.legend(bbox_to_anchor=(1.04,0.5), loc="center left", borderaxespad=0)

ax.set_title("Imports by country")
plt.show()

```

```python
fig, ax = plt.subplots()

ax = df_outflows_monthly.plot.bar(stacked=True, figsize=(16,8), ax=ax, cmap='tab20')

ax.set_xticklabels([x.strftime("%Y-%m") for x in df_outflows_monthly.index])

ax.legend(bbox_to_anchor=(1.04,0.5), loc="center left", borderaxespad=0)
ax.set_title("Exports by country")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,5)) # Create matplotlib figure

 #= fig.add_subplot() # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.

width = 0.2

df_flows_monthly['Germany Export(TWh)'].plot(kind='bar',  ax=ax, width=width, position=1, label='Export', color='green')
df_flows_monthly['Germany Import(TWh)'].plot(kind='bar',  ax=ax2, width=width, position=0, label='Import', color='red')

ax.set_ylabel('TWh')


ax.legend(bbox_to_anchor=(1.04,0.5), loc="center left", borderaxespad=0)
ax2.legend(bbox_to_anchor=(1.04,0.6), loc="center left", borderaxespad=0)
plt.title('Total export and import 2020')
plt.show()
#ax.set_xticklabels([x.strftime("%Y-%m") for x in df_inflows_monthly.index])


```

```python
fig, ax = plt.subplots(figsize=(18, 3))


# define the shift
shift = pd.to_timedelta('5D')
width=5
# modify the base of each columns, can do with a for loop
ax.bar(gener_monthly.index, gener_monthly['Total Generation(TWh)'], width, color='green', label='Total generation') 
ax.bar(df_load_monthly.index + shift, df_load_monthly["Actual Total Load(TWh)"], width, color='red', label='Total load')
ax.bar(df_flows_monthly['Germany Import(TWh)'].index - shift, df_flows_monthly['Germany Import(TWh)'], width, color='blue', label='Total imports')
ax.bar(df_flows_monthly['Germany Export(TWh)'].index + 2*shift, df_flows_monthly['Germany Export(TWh)'], width, color='orange', label='Total exports')

#ax.bar(testdata.index - shift, testdata["C"])
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
ax.set_ylabel('TWh')
plt.title('Monthly comparison of generation, load, imports and exports')
plt.legend()
plt.show()
```
