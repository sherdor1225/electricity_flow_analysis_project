```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Cross_Border_totalflow_cleaned.xlsx')
df_new = pd.Series(pd.date_range(start='1/1/2020 00:00:00', end='31/12/2020 23:00:00', freq='1h'))
df1 = pd.concat([df_new, df], axis=1)
df1.drop(columns=['Time'], inplace=True)
df1.rename(columns={0: 'Timestep'}, inplace=True)
df1['Timestep']=pd.to_datetime(df1['Timestep'])
df1.set_index('Timestep', inplace=True)
df1.head()
```

```python
hourly_mean=df1.groupby(df1.index.hour).mean()*1000
```

```python
hourly_mean_im = hourly_mean.iloc[:, 0:21:2]
hourly_mean_ex = hourly_mean.iloc[:, 1:22:2]
```

```python
hourly_mean_im['Total'] = hourly_mean['Germany Import(TWh)']
hourly_mean_ex['Total'] = hourly_mean['Germany Export(TWh)']
```

```python
hourly_mean_ex.head()
```

```python
hourly_mean_im.head()
```

```python
fig_15 = plt.figure(figsize=(18,8))

axes_15 = fig_15.add_axes([0.0, 0.0, 0.9, 0.9])




sns.histplot(hourly_mean_ex,  x = hourly_mean_ex['Total'], kde=True)


# axes_14.set_xticks(np.arange(2.5, 6, 1))
sns.set_style("darkgrid")
# sns.set_context("paper", font_scale=1.5, rc={"lines.linewidth": 2.5})
```

```python
fig_14 = plt.figure(figsize=(18,8))

axes_14 = fig_14.add_axes([0.0, 0.0, 0.9, 0.9])




sns.histplot(hourly_mean_im, x = hourly_mean_im['Total'], bins= 20)


# axes_14.set_xticks(np.arange(2.5, 6, 1))
sns.set_style("darkgrid")
sns.set_context("paper", font_scale=1.5, rc={"lines.linewidth": 2.5})
```

```python
fig_13 = plt.figure(figsize=(12,6))

axes_13 = fig_13.add_axes([0.0, 0.0, 0.9, 0.9])




sns.histplot(hourly_mean_im, x = hourly_mean_im['France (FR) > Germany (DE)(TWh)'], bins=15, kde=True, color='r')
axes_13.set_title('Yearly Cross-border import by the hour of the day from FR to DE')
axes_13.set_xlabel('FR > DE (GWh)')

# axes_14.set_xticks(np.arange(2.5, 6, 1))
sns.set_style("darkgrid")
sns.set_context("paper", font_scale=1.5, rc={"lines.linewidth": 2.5})
```

```python
fig_13 = plt.figure(figsize=(12,6))

axes_13 = fig_13.add_axes([0.0, 0.0, 0.9, 0.9])




sns.histplot(hourly_mean_ex, x = hourly_mean_ex['Germany (DE) > France (FR)(TWh)'], bins=15, kde=True, color='green')
axes_13.set_xlabel('DE > FR (GWh)')
axes_13.set_title('Yearly Cross-border export by the hour of the day from DE to FR')
# axes_14.set_xticks(np.arange(2.5, 6, 1))
sns.set_style("darkgrid")
sns.set_context("paper", font_scale=1.5, rc={"lines.linewidth": 2.5})
```
