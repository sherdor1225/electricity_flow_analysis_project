```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Cross_Border_totalflow_cleaned.xlsx')
df_new = pd.Series(pd.date_range(start='1/1/2020 00:00:00', end='31/12/2020 23:00:00', freq='1h'))
df1 = pd.concat([df_new, df], axis=1)
df1.drop(columns=['Time'], inplace=True)
df1.rename(columns={0: 'Timestep'}, inplace=True)
df1['Timestep']=pd.to_datetime(df1['Timestep'])
df1.set_index('Timestep', inplace=True)
df1.head()
```

```python
df1=df1*1000
df1_mean=df1.resample('D').mean()
df1_std=df1.resample('D').std()
```

```python
df1_mean.reset_index(inplace=True)
```

```python
merged_inner_ex_mean = pd.merge(left=df1_mean, right=export_days, left_on='Timestep', right_on='Timestep')
```

```python
merged_inner_im_mean = pd.merge(left=df1_mean, right=import_days, left_on='Timestep', right_on='Timestep')
```

```python
merged_inner_im_std = pd.merge(left=df1_std, right=import_days, left_on='Timestep', right_on='Timestep')
merged_inner_ex_std = pd.merge(left=df1_std, right=export_days, left_on='Timestep', right_on='Timestep')
```

```python
merged_inner_ex_mean.set_index('Timestep', inplace=True)
merged_inner_im_mean.set_index('Timestep', inplace=True)
merged_inner_ex_std.set_index('Timestep', inplace=True)
merged_inner_im_std.set_index('Timestep', inplace=True)
```

```python
fig_2 = plt.figure(figsize=(18,12))

axes_2 = fig_2.add_axes([0.0, 0.0, 0.9, 0.9])

axes_2.plot(merged_inner_ex_mean.index, merged_inner_ex_mean['Germany Export(TWh)'], 'b-', label='mean_export')
axes_2.fill_between(merged_inner_ex_mean.index, merged_inner_ex_mean['Germany Export(TWh)'] - 
                 merged_inner_ex_std['Germany Export(TWh)'], 
                 merged_inner_ex_mean['Germany Export(TWh)'] + 
                 merged_inner_ex_std['Germany Export(TWh)'], color='b', alpha=0.2)

axes_2.plot(merged_inner_im_mean.index, merged_inner_im_mean['Germany Import(TWh)'], 'r-', label='mean_import')
axes_2.fill_between(merged_inner_im_mean.index, merged_inner_im_mean['Germany Import(TWh)'] - 
                 merged_inner_im_std['Germany Import(TWh)'], 
                 merged_inner_im_mean['Germany Import(TWh)'] + 
                 merged_inner_im_std['Germany Import(TWh)'], color='r', alpha=0.2)



axes_2.set_ylabel('Cross-border flow, GWh')
axes_2.set_title('Cross-border export/import total')
axes_2.legend(loc=0)
sns.set_style("darkgrid")
sns.set_context("paper", font_scale=1.5, rc={"lines.linewidth": 2.5})
```

```python
fig_3 = plt.figure(figsize=(8,5), dpi=100)

axes_3 = fig_3.add_axes([0.0, 0.0, 0.9, 0.9])

axes_3.plot(merged_inner_ex_mean.index, merged_inner_ex_mean['Germany (DE) > Austria (AT)(TWh)'], 'b-', label='mean_DE>AT')
axes_3.fill_between(merged_inner_ex_mean.index, merged_inner_ex_mean['Germany (DE) > Austria (AT)(TWh)'] - 
                 merged_inner_ex_std['Germany (DE) > Austria (AT)(TWh)'], 
                 merged_inner_ex_mean['Germany (DE) > Austria (AT)(TWh)'] + 
                 merged_inner_ex_std['Germany (DE) > Austria (AT)(TWh)'], color='b', alpha=0.2)

axes_3.set_ylabel('Cross-border flow, GWh')
axes_3.set_title('Cross-border export from DE to AT')
axes_3.legend(loc=0)
sns.set_style("darkgrid")
sns.set_context("paper", font_scale=1.5, rc={"lines.linewidth": 2.5})
```

```python
df_ger= pd.read_excel('data_hourly_incase_import.xlsx')
df_ger_new = pd.Series(pd.date_range(start='1/1/2020 00:00:00', end='31/12/2020 23:00:00', freq='1h'))
df1_ger = pd.concat([df_ger_new, df_ger], axis=1)
df1_ger.drop(columns=['Timestep'], inplace=True)
df1_ger.rename(columns={0: 'Timestep'}, inplace=True)
df1_ger['Timestep']=pd.to_datetime(df1_ger['Timestep'])
df1_ger.set_index('Timestep', inplace=True)
df1_ger.head()
```

```python
df1_ger_mean=df1_ger.resample('D').mean()
df1_ger_mean.reset_index(inplace=True)
df1_ger_mean.head()
```

```python
df1_ger_exporter_mean = pd.merge(left=df1_ger_mean, right=export_days, left_on='Timestep', right_on='Timestep')
df1_ger_importer_mean = pd.merge(left=df1_ger_mean, right=import_days, left_on='Timestep', right_on='Timestep')
df1_ger_exporter_mean.drop(df1_ger_exporter_mean.columns[16], axis=1, inplace=True)
df1_ger_importer_mean.drop(df1_ger_importer_mean.columns[16], axis=1, inplace=True)
df1_ger_exporter_mean.set_index('Timestep', inplace=True)
df1_ger_importer_mean.set_index('Timestep', inplace=True)
```

```python
# fig_4 = plt.figure(figsize=(8,5), dpi=100)

# axes_4 = fig_4.add_axes([0.0, 0.0, 0.9, 0.9])
sns.set_style("white")
axes_4 = df1_ger_exporter_mean.plot.bar(stacked=True, figsize=(12,8), width=0.9, colormap='rainbow')

monthly_timestamps = [timestamp for idx, timestamp in enumerate(df1_ger_exporter_mean.index)
                      if (timestamp.month != df1_ger_exporter_mean.index[idx-1].month) | (idx == 0)]

step = 1
while len(monthly_timestamps[::step]) > 12: # increase number if time range >3 years
    step += 1
timestamps = monthly_timestamps[::step]

# Create tick labels from timestamps
labels = [ts.strftime('%b\n%Y') if ts.year != timestamps[idx-1].year
          else ts.strftime('%b') for idx, ts in enumerate(timestamps)]

# Set major ticks and labels
axes_4.set_xticks([df1_ger_exporter_mean.index.get_loc(ts) for ts in timestamps])
axes_4.set_xticklabels(labels)

# Set minor ticks without labels
axes_4.set_xticks([df1_ger_exporter_mean.index.get_loc(ts) for ts in monthly_timestamps], minor=True)


axes_4.set_ylabel('Energy Generation, TWh')
axes_4.set_title('Energy Generation per type when exporting')
axes_4.legend(loc=0)
axes_4.legend(bbox_to_anchor=(1.04,0.5), loc="center left", borderaxespad=0)

sns.set_context("talk", font_scale=1.5, rc={"lines.linewidth": 2.5})
```

```python
# fig_4 = plt.figure(figsize=(8,5), dpi=100)

# axes_4 = fig_4.add_axes([0.0, 0.0, 0.9, 0.9])
sns.set_style("white")
axes_4 = df1_ger_importer_mean.plot.bar(stacked=True, figsize=(12,8), width=0.9, colormap='rainbow')

monthly_timestamps = [timestamp for idx, timestamp in enumerate(df1_ger_importer_mean.index)
                      if (timestamp.month != df1_ger_importer_mean.index[idx-1].month) | (idx == 0)]

step = 1
while len(monthly_timestamps[::step]) > 12: # increase number if time range >3 years
    step += 1
timestamps = monthly_timestamps[::step]

# Create tick labels from timestamps
labels = [ts.strftime('%b\n%Y') if ts.year != timestamps[idx-1].year
          else ts.strftime('%b') for idx, ts in enumerate(timestamps)]

# Set major ticks and labels
axes_4.set_xticks([df1_ger_importer_mean.index.get_loc(ts) for ts in timestamps])
axes_4.set_xticklabels(labels)

# Set minor ticks without labels
axes_4.set_xticks([df1_ger_importer_mean.index.get_loc(ts) for ts in monthly_timestamps], minor=True)


axes_4.set_ylabel('Energy Generation, TWh')
axes_4.set_title('Energy Generation per type when importing')
axes_4.legend(loc=0)
axes_4.legend(bbox_to_anchor=(1.04,0.5), loc="center left", borderaxespad=0)

sns.set_context("talk", font_scale=1.5, rc={"lines.linewidth": 2.5})
```

```python
fig_5 = plt.figure()

axes_5 = fig_5.add_axes([0.0, 0.0, 0.9, 0.9])

axes_5 = df1_ger_exporter_mean.resample('M').mean().plot.bar(stacked=True, figsize=(16,5), colormap='rainbow', ax=axes_5)

axes_5.set_xticklabels([x.strftime("%Y-%m") for x in df1_ger_exporter_mean.resample('M').mean().index])

axes_5.legend(bbox_to_anchor=(1.04,0.5), loc="center left", borderaxespad=0)
```
