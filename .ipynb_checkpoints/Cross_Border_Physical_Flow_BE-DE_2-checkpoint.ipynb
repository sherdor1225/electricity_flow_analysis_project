```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Cross_Border_Physical_Flow_BE-DE_modified_1.xlsx', index_col=[0])
df.head()
```

```python
df.tail()
```

```python
df['Time']=pd.to_datetime(df['Time'])
```

```python
df['Time'] = df['Time'].dt.round('T')
```

```python
df.set_index(['Time'], inplace=True)
```

```python
df.head(30)
```

```python
data_hourly=df.resample('H').sum()
```

```python
data_hourly=data_hourly/(4*1000000)
```

```python
data_hourly=data_hourly.add_suffix('(TWh)')
```

```python
data_hourly
```

```python
data_summer_day = data_hourly.loc['2020-06-01':'2020-07-01']
```

```python
# Create pandas stacked bar chart
ax = data_hourly.resample('d').sum().plot.bar(stacked=True, figsize=(16,5))

# Create list of monthly timestamps by selecting the first weekly timestamp of each
# month (in this example, the first Sunday of each month)
monthly_timestamps = [timestamp for idx, timestamp in enumerate(data_hourly.resample('d').sum().index)
                      if (timestamp.month != data_hourly.resample('d').sum().index[idx-1].month) | (idx == 0)]

# Automatically select appropriate number of timestamps so that x-axis does
# not get overcrowded with tick labels
step = 1
while len(monthly_timestamps[::step]) > 12: # increase number if time range >3 years
    step += 1
timestamps = monthly_timestamps[::step]

# Create tick labels from timestamps
labels = [ts.strftime('%b\n%Y') if ts.year != timestamps[idx-1].year
          else ts.strftime('%b') for idx, ts in enumerate(timestamps)]

# Set major ticks and labels
ax.set_xticks([data_hourly.resample('d').sum().index.get_loc(ts) for ts in timestamps])
ax.set_xticklabels(labels)

# Set minor ticks without labels
ax.set_xticks([data_hourly.resample('d').sum().index.get_loc(ts) for ts in monthly_timestamps], minor=True)

# Rotate and center labels
ax.figure.autofmt_xdate(rotation=0, ha='center')

ax.legend(bbox_to_anchor=(1.04,0.5), loc="center left", borderaxespad=0)
plt.show()
```
