```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Cross_Border_totalflow_cleaned.xlsx')
df.head()
```

```python
df_new = pd.Series(pd.date_range(start='1/1/2020 00:00:00', end='31/12/2020 23:00:00', freq='1h'))
df_new
```

```python
df1 = pd.concat([df_new, df], axis=1)
df1.drop(columns=['Time'], inplace=True)
df1.rename(columns={0: 'Timestep'}, inplace=True)
df1['Timestep']=pd.to_datetime(df1['Timestep'])
df1.set_index('Timestep', inplace=True)
```

```python
df1.head()
```

```python
hourly_mean=df1.groupby(df1.index.hour).mean()*1000
hourly_std=df1.groupby(df1.index.hour).std()*1000
```

```python
hourly_mean_im = hourly_mean.iloc[:, 0:21:2]
hourly_mean_ex = hourly_mean.iloc[:, 1:22:2]

hourly_std_im = hourly_std.iloc[:, 0:21:2]
hourly_std_ex = hourly_std.iloc[:, 1:22:2]
```

```python
fig_0 = plt.figure(figsize=(18,8))

axes_0 = fig_0.add_axes([0.0, 0.0, 0.9, 0.9])

axes_0.plot(hourly_mean_im.index, hourly_mean_im['Austria (AT) > Germany (DE)(TWh)'], 'r-', label='mean_AT>DE',
           marker ='o', markersize=8, markerfacecolor='red')
axes_0.fill_between(hourly_mean_im.index, hourly_mean_im['Austria (AT) > Germany (DE)(TWh)'] - 
                 hourly_std_im['Austria (AT) > Germany (DE)(TWh)'], 
                 hourly_mean_im['Austria (AT) > Germany (DE)(TWh)'] + 
                 hourly_std_im['Austria (AT) > Germany (DE)(TWh)'], color='r', alpha=0.2)
axes_0.set_xticks(hourly_mean_im.index)
axes_0.set_ylabel('Cross-border flow, GWh')
axes_0.set_xlabel('Hour of the day')
axes_0.set_title('Average Cross-border import from AT to DE by the hour of the day')
axes_0.legend(loc=0)
sns.set_style("darkgrid")
sns.set_context("paper", font_scale=1.5, rc={"lines.linewidth": 2.5})
```

```python
fig_1 = plt.figure(figsize=(18,8))

axes_1 = fig_1.add_axes([0.0, 0.0, 0.9, 0.9])

axes_1.plot(hourly_mean_ex.index, hourly_mean_ex['Germany (DE) > Austria (AT)(TWh)'], 'b-', label='mean_DE>AT',
           marker ='o', markersize=8, markerfacecolor='blue')
axes_1.fill_between(hourly_mean_ex.index, hourly_mean_ex['Germany (DE) > Austria (AT)(TWh)'] - 
                 hourly_std_ex['Germany (DE) > Austria (AT)(TWh)'], 
                 hourly_mean_ex['Germany (DE) > Austria (AT)(TWh)'] + 
                 hourly_std_ex['Germany (DE) > Austria (AT)(TWh)'], color='b', alpha=0.2)
axes_1.set_xticks(hourly_mean_ex.index)
axes_1.set_ylabel('Cross-border flow, GWh')
axes_1.set_xlabel('Hour of the day')
axes_1.set_title('Average Cross-border export from AT to DE by the hour of the day')
axes_1.legend(loc=0)
sns.set_style("darkgrid")
sns.set_context("paper", font_scale=1.5, rc={"lines.linewidth": 2.5})
```

```python
fig_3 = plt.figure(figsize=(18,12))

axes_3 = fig_3.add_axes([0.0, 0.0, 0.9, 0.9])


sns.lineplot(data=hourly_mean_ex, markers=False, dashes=False, palette="icefire")

axes_3.set_ylabel('Cross-border flow, GWh')
axes_3.set_title('Yearly Cross-border export by the hour of the day')
axes_3.legend(loc=0)
sns.set_style("darkgrid")
sns.set_context("talk", font_scale=1.5, rc={"lines.linewidth": 2.5})
plt.legend(bbox_to_anchor=(1.04,0.5), loc="center left", borderaxespad=0)
```
