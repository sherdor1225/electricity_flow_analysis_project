```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Actual_Generation_per_Production_Type_modified_copy.xlsx', index_col=[0])
df.head()
```

```python
df['Date and Time']=pd.to_datetime(df['Date and Time'])
```

```python
df['Date and Time'] = df['Date and Time'].dt.round('T')
```

```python
df.head(15)
```

```python
df['Biomass'].sum()/(4*1000000)
```

```python
df.set_index(['Date and Time'], inplace=True)
```

```python
df
```

```python
hourly_data = df.resample('H').sum()
```

```python
hourly_data
```

```python
hourly_data=hourly_data/(4*1000000)
```

```python
hourly_data=hourly_data.add_suffix(' (TWh)')
```

```python
hourly_data.head(15)
```

```python
total_by_type=hourly_data.sum()
```

```python
total_by_type=total_by_type.reset_index().transpose()
```

```python
total_by_type.reset_index(drop=True, inplace = True)
```

```python
total_by_type.rename(columns=total_by_type.iloc[0], inplace=True)
```

```python
total_by_type.drop(index=0, inplace=True)
total_by_type.reset_index(drop=True, inplace=True)
```

```python
total_by_type
```

```python
plt.figure(figsize=(18, 12))


plt.title('Total Production in 2020, TWh', fontdict = {'fontweight': 'bold', 'fontsize': 18})

types = ['Biomass (TWh)', 'Fossil Brown coal/Lignite (TWh)', 'Fossil Gas (TWh)',
       'Fossil Hard coal (TWh)', 'Fossil Oil (TWh)', 'Geothermal (TWh)',
       'Hydro Pumped Storage (TWh)', 'Hydro Run-of-river and poundage (TWh)',
       'Hydro Water Reservoir (TWh)', 'Nuclear (TWh)', 'Other (TWh)',
       'Other renewable (TWh)', 'Solar (TWh)', 'Waste (TWh)',
       'Wind Offshore (TWh)', 'Wind Onshore (TWh)']

def absolute_value(val):
    a  = np.round(val/100.*total_by_type.loc[0].sum(), 0)
    return a


plt.pie(total_by_type.loc[0], labels = types, autopct = absolute_value, pctdistance = 0.9, explode = (0, 0, 0, 0, 0.4,
                                                                                                     0.4, 0, 0, 0.4, 0,
                                                                                                      0.4, 0.4, 0, 0.2, 0, 0,))



plt.tight_layout()
plt.show()


```
