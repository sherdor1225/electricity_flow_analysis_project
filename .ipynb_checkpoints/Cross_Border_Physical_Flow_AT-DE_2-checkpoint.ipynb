```python
de_hourly_halfyear = de_hourly_export['2020-01-01':'2020-07-01']
at_hourly_halfyear = at_hourly_export['2020-01-01':'2020-07-01']
```

```python
de_hy=df['Germany (DE) > Austria (AT)']['2020-01-01':'2020-07-01']
at_hy=df['Austria (AT) > Germany (DE)']['2020-01-01':'2020-07-01']
```

```python
plt.figure(figsize=(18, 8))

plt.subplot(2, 1, 1)
plt.title('First half of the year', fontdict = {'fontweight': 'bold', 'fontsize': 18})
plt.bar(df.loc['2020-01-01':'2020-07-01'].index , de_hy, width=0.3, color = 'red',label='Germany (DE) > Austria (AT)')
plt.legend(loc='upper center')

plt.subplot(2, 1, 2)
plt.bar(df.loc['2020-01-01':'2020-07-01'].index , at_hy, width=0.3, label='Austria (AT) > Germany (DE)')
plt.legend(loc='upper center')

plt.tight_layout()
plt.show()
```

```python
de_second_hy=df['Germany (DE) > Austria (AT)']['2020-07-01':'2020-12-31']
at_second_hy=df['Austria (AT) > Germany (DE)']['2020-07-01':'2020-12-31']
```

```python
plt.figure(figsize=(18, 8))



plt.subplot(2, 1, 1)
plt.title('Second half of the year', fontdict = {'fontweight': 'bold', 'fontsize': 18})
plt.bar(df.loc['2020-07-01':'2020-12-31'].index , de_second_hy, width=0.3, color = 'red',label='Germany (DE) > Austria (AT)')
plt.legend(loc='upper center')

plt.subplot(2, 1, 2)
plt.bar(df.loc['2020-07-01':'2020-12-31'].index , at_second_hy, width=0.3, label='Austria (AT) > Germany (DE)')
plt.legend(loc='upper center')

plt.tight_layout()
plt.show()
```
